```python
# Exercise 15: Analysis of the Boston Dataset
import pandas as pd
import numpy as np
import statsmodels.api as sm
import matplotlib.pyplot as plt
import seaborn as sns

# Boston dataset 
boston_df = pd.read_csv('Boston.csv')

# Inspecting the dataset
print(boston_df.head())
print(boston_df.info())

# Define the response variable (crim - per capita crime rate)
y = boston_df['crim']

# list of predictor variables (all columns except 'crim')
X = boston_df.drop(columns=['crim'])

### Part (a): Simple Linear Regression for Each Predictor ###

# Simple linear 
predictors = X.columns
significant_predictors = []

for predictor in predictors:
    X_single = sm.add_constant(boston_df[predictor])  # Add intercept
    model = sm.OLS(y, X_single).fit()  # Fit the model
    print(f"\nSimple linear regression for {predictor}:")
    print(model.summary())
    
    # Check if the p-value for the predictor is significant 
    if model.pvalues[predictor] < 0.05:
        significant_predictors.append(predictor)

# Print significant predictors
print(f"\nSignificant predictors from simple linear regression (p < 0.05): {significant_predictors}")

# Visualizing the relationships between 'crim' and each predictor
for predictor in predictors:
    plt.figure(figsize=(6, 4))
    sns.regplot(x=boston_df[predictor], y=boston_df['crim'])
    plt.title(f"Crime Rate vs {predictor}")
    plt.xlabel(predictor)
    plt.ylabel('Crime Rate (CRIM)')
    plt.show()

### Part (b): Multiple Regression Model ###

# constant to the predictors
X_with_constant = sm.add_constant(X)

# multiple linear regression model
multiple_model = sm.OLS(y, X_with_constant).fit()

# the summary of the model
print("\nMultiple Regression Model:")
print(multiple_model.summary())

significant_multiple = multiple_model.pvalues[multiple_model.pvalues < 0.05].index.tolist()
print(f"\nSignificant predictors in the multiple regression model (p < 0.05): {significant_multiple}")

### Part (c): Compare Results from Simple and Multiple Regression ###

# Coefficients from simple linear regressions
simple_coefs = []

for predictor in predictors:
    X_single = sm.add_constant(boston_df[predictor])
    model = sm.OLS(y, X_single).fit()
    simple_coefs.append(model.params[predictor])

# Coefficients from the multiple regression model
multiple_coefs = multiple_model.params[predictors]

# Plot the comparison of coefficients
plt.figure(figsize=(8, 6))
plt.scatter(simple_coefs, multiple_coefs)
plt.xlabel('Simple Linear Regression Coefficients')
plt.ylabel('Multiple Linear Regression Coefficients')
plt.title('Comparison of Coefficients: Simple vs. Multiple Regression')
plt.axhline(0, color='r', linestyle='--')
plt.axvline(0, color='r', linestyle='--')
plt.grid(True)
plt.show()

### Part (d): Check for Non-Linear Relationships ###

# Checking for non-linear relationships by fitting polynomial terms (quadratic and cubic)
for predictor in predictors:
    # Add quadratic and cubic terms for the predictor
    X_poly = pd.DataFrame()
    X_poly[f'{predictor}'] = boston_df[predictor]
    X_poly[f'{predictor}^2'] = boston_df[predictor] ** 2
    X_poly[f'{predictor}^3'] = boston_df[predictor] ** 3
    X_poly = sm.add_constant(X_poly)

    # Fit the model with polynomial terms
    poly_model = sm.OLS(y, X_poly).fit()

    # Print the summary
    print(f"\nPolynomial regression for {predictor}:")
    print(poly_model.summary())

    # Plot the polynomial regression
    sns.regplot(x=boston_df[predictor], y=boston_df['crim'], order=3, line_kws={"color": "red"})
    plt.title(f"Polynomial Fit for {predictor}")
    plt.xlabel(predictor)
    plt.ylabel('Crime Rate (CRIM)')
    plt.show()
```
